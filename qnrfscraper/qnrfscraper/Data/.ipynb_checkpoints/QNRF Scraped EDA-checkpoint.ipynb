```python
def process_research_area(text_array):
    """
    Processes the research_table_info containing research (Research Area,Sub Research Area,Sub Speciality,Primary?,Secondary?)
    Strips \n \t and spaces
    :param text_array:
    :return: a list containing the processed departments in this order "research_area_1,sub_category_1,speciality_1,research_area_2,sub_category_2,speciality_2"
    """
    table_labels = ["research_area_1,sub_category_1,speciality_1,research_area_2,sub_category_2,speciality_2"]
    text_array = np.array(text_array)
    text_array = text_array.reshape(-1,5)
    text_array = text_array[:2]
    text_array = np.delete(text_array,[3,4],1)
    text_array = np.array([[item.strip() for item in x] for x in text_array])
    text_array = text_array.flatten()
    print(text_array)
    
    while len(text_array) < 6:
        text_array = np.append(text_array,"")
    return text_array
```

```python
# Define an annotation-generating function
def make_annotation(x, y):
    return Annotation(
        text=str(x),     # text is the y-coord
        showarrow=False, # annotation w/o arrows, default is True
        x=1,               # set x position
        xref='x',          # position text horizontally with x-coords
        xanchor='left',  # x position corresp. to center of text
        yref='y',            # set y position
        yanchor='auto',       # position text vertically with y-coords
        y=x,                 # y position corresp. to top of text
        textangle=0,       # Rotating the text upwards
        font=annotation.Font(
            color='#262626',  # set font color
            size=13           #   and size
        )
    )

```
